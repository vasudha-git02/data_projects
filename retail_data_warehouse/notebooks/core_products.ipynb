```python
from pyspark.sql import SparkSession
from pyspark.sql import functions as f
from pyspark.sql.functions import col
import os
```

```python
spark = SparkSession.builder.appName("Create Core Products").getOrCreate()
```

```python
base_path = r"C:\Users\vasudha.tanniru\Documents\GitHub\data_projects\retail_data_warehouse\warehouse"
staging_path = os.path.join(base_path,"staging")
core_path = os.path.join(base_path,"core")
```

```python
products_df = spark.read.parquet(os.path.join(staging_path,"products"))
```

```python
products_df.show()
```

```python
core_products = (
    products_df
    .withColumn("product_category_name", f.trim(f.col("product_category_name")))
    .dropDuplicates(["product_id"])
)
```

```python
core_products = core_products.filter(f.col("product_id").isNotNull())
```

```python
core_products.write.mode("overwrite").parquet(os.path.join(core_path, "core_products"))
```
