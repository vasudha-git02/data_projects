```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import trim, col
```

```python
spark = SparkSession.builder.appName("CsvToParquet").getOrCreate()
```

```python
customerDF = spark.read.csv(r"C:\Users\vasudha.tanniru\Documents\GitHub\data_projects\retail_data_warehouse\datasets\olist_customers_dataset.csv",header=True,inferSchema=True)
```

```python
customerDF.show()
```

```python
for old_col in customerDF.columns:
    customerDF = customerDF.withColumnRenamed(old_col,old_col.strip().lower())
```

```python
customerDF = customerDF.drop_duplicates()
```

```python
string_cols = [f.name for f in customerDF.schema.fields if f.dataType.simpleString() == "string"]
for colname in string_cols:
    customerDF = customerDF.withColumn(colname, trim(col(colname)))
    
```

```python
customerDF.coalesce(1).write.mode("overwrite").parquet(
    r"C:\Users\vasudha.tanniru\Documents\GitHub\data_projects\retail_data_warehouse\warehouse\staging\customers"
)
```
