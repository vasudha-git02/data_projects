```python
import pandas as pd
import numpy as np
import datetime as dt
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
os.getcwd()
```

```python
df = pd.read_csv("full.csv")
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['bathrooms_text'] = df['bathrooms_text'].fillna("Unknown")
```

```python
df.head()
```

```python
df.info()
```

```python
print(df['bedrooms'].median())
```

```python
df['bedrooms'] = df['bedrooms'].fillna(df['bedrooms'].median())
```

```python
df.isnull().sum()
```

```python
print(df['beds'].median())
```

```python
df['beds'] = df['beds'].fillna(df['beds'].median())
```

```python
df.head()
```

```python
df.duplicated().sum()
```

```python
df['price_per_bedroom'] = df['price']/df['bedrooms']
```

```python
df.info()
```

```python
print(df['price'].dtype)
```

```python
# Since price has $, we need to remove it before converting it into a integer kind column
df['price'] = df['price'].replace({'\$':'',',':''},regex=True).astype(float)
```

```python
df.head()
```

```python
df['price_per_bedroom'] = df['price']/df['beds']
```

```python
df.head(30)
```

```python
# Luxury vs budget listing
df['listing_category'] = df['price'].apply(lambda x:'Luxury' if x > 500.0 else 'Budget' if x < 100.0 else 'Standard')
```

```python
df.head()
```

```python
print(df['listing_category'].value_counts())
```

```python
df['price_per_bed'] = df.apply(lambda row: row['price'] / row['beds'] 
                               if row['beds'] > 0 else row['price'], axis=1)
```

```python
df.head()
```

```python
df['price_per_accomodate'] = df.apply(lambda row:row['price']/row['accommodates'] if row['accommodates'] > 0 else row['price'],axis = 1)
```

```python
df.head()
```

```python
df['availability_rate'] = round((df['availability_365']/365)*100,2)
```

```python
df.head()
```

```python
df['num_amenities'] = df['amenities'].apply(lambda x: len(x.split(',')) if pd.notnull(x) else 0)
```

```python
df.head()
```

```python
df['bedroom_to_accommodate_ratio'] = df['bedrooms'] / df['accommodates']
```

```python
df.head()
```

```python
print(df['price'].describe())
```

```python
# Price vs number of amenities
plt.figure(figsize = (8,5))
sns.boxplot(x=df['availability_365'], y = df['price'])
plt.title("Price vs Availability")
plt.xlabel("Availability (days per year)")
plt.ylabel("Price ($)")
plt.ylim(0, df['price'].quantile(0.95)) # Limiting exteme outliers
plt.show()
```
